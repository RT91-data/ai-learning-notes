## Parameters vs Hyperparameters

**Parameters** are values the model learns automatically during training.  
**Hyperparameters** are values YOU set before training begins.

## How it actually works

During training, the model adjusts its parameters thousands of times to reduce error.  
Hyperparameters control *how* that adjustment happens — you choose them upfront.  
Wrong hyperparameters = slow training or a model that never learns properly.

## Where it breaks or gets dangerous

Choosing a learning rate (hyperparameter) too high = model overshoots, never converges.  
Choosing too low = training takes forever or gets stuck.  
There is no perfect formula — hyperparameter tuning is trial and error.

```python
# Parameters vs Hyperparameters — simple illustration

# Hyperparameter — YOU set this before training
learning_rate = 0.01

# Simulated parameters — model learns these during training
import random
weights = [random.uniform(-1, 1) for _ in range(5)]

print(f"Hyperparameter (learning rate): {learning_rate}")
print(f"Parameters (weights before training): {weights}")
```

In D365, workflow rules (approval thresholds, routing conditions) are like hyperparameters —  
a human configures them before the system runs.  
The actual transaction data the system processes and stores is like parameters —
