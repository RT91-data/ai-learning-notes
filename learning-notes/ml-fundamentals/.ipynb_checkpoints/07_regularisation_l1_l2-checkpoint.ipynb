## Regularisation — L1 and L2

Regularisation prevents overfitting by adding a penalty to the cost function for large weights.  
Forces the model to stay simple. Two main types: L1 (Lasso) and L2 (Ridge).

## How it actually works

**Normal cost function:**  
Cost = how wrong the model is on training data

**With regularisation:**  
Cost = how wrong the model is + penalty for large weights

The model now has two jobs simultaneously:  
1. Get predictions right  
2. Keep weights small  

That tension prevents memorisation of training data.

**L1 Regularisation (Lasso):**  
Penalty = sum of absolute values of all weights.  
Effect: pushes many weights to exactly zero.  
Removes irrelevant features from the model entirely.  
Use when: you suspect many input features are irrelevant.

**L2 Regularisation (Ridge):**  
Penalty = sum of squared values of all weights.  
Effect: shrinks all weights toward zero but rarely to exactly zero.  
Keeps all features but reduces their individual influence.  
Use when: all features are somewhat relevant but none should dominate.

**Lambda (λ) — the regularisation hyperparameter:**  
Controls how strong the penalty is.  
High λ = strong penalty = simpler model = more bias, less variance.  
Low λ = weak penalty = complex model = less bias, more variance.  
You're always back to the bias-variance tradeoff.

## Where it breaks or gets dangerous

**Lambda too high:**  
Penalty so strong the model becomes too simple.  
Underfitting. High bias.  
You've overcorrected from overfitting into underfitting.

**Lambda too low:**  
Penalty too weak to make a difference.  
Model still overfits. Regularisation did nothing useful.

**Wrong type of regularisation:**  
Using L2 when most features are irrelevant — model keeps useless features with small weights.  
Using L1 when all features matter — model throws away useful information.

**The core danger:**  
Regularisation requires tuning lambda correctly.  
Too much or too little both produce bad models.  
Always evaluate on a validation set when tuning lambda.

```python
# L1 vs L2 Regularisation — conceptual illustration

def l1_penalty(weights, lambda_val):
    """L1: sum of absolute values of weights"""
    return lambda_val * sum(abs(w) for w in weights)

def l2_penalty(weights, lambda_val):
    """L2: sum of squared values of weights"""
    return lambda_val * sum(w**2 for w in weights)

def total_cost(base_cost, weights, lambda_val, regularisation_type):
    if regularisation_type == "L1":
        penalty = l1_penalty(weights, lambda_val)
    elif regularisation_type == "L2":
        penalty = l2_penalty(weights, lambda_val)
    else:
        penalty = 0
    return base_cost + penalty

# Scenario: model with some large and some small weights
weights_overfitted = [15.2, 0.1, 12.8, 0.05, 9.3, 0.02, 11.1]
base_cost = 0.15
lambda_val = 0.1

print("Weights before regularisation:", [round(w, 2) for w in weights_overfitted])
print()

no_reg = total_cost(base_cost, weights_overfitted, 0, "none")
l1_cost = total_cost(base_cost, weights_overfitted, lambda_val, "L1")
l2_cost = total_cost(base_cost, weights_overfitted, lambda_val, "L2")

print(f"Base cost (no regularisation): {no_reg:.4f}")
print(f"Total cost with L1:            {l1_cost:.4f}  → penalises large weights, drives small ones to zero")
print(f"Total cost with L2:            {l2_cost:.4f}  → penalises large weights, shrinks all")
print()
print("L1 effect: irrelevant features (near-zero weights) get eliminated entirely")
print("L2 effect: all features kept but large weights are shrunk significantly")
```

## My D365 analogy

In D365, financial reporting often has 50+ dimensions configured.  
Not all of them matter for every report or analysis.

**L1 regularisation** is like a D365 architect reviewing all 50 dimensions  
and saying: only 8 of these actually drive meaningful analysis.  
Remove the other 42 from the reporting structure entirely.  
Cleaner, faster, more maintainable.

**L2 regularisation** is like keeping all 50 dimensions  
but reducing the weight given to each one in aggregated reports.  
No single dimension dominates the analysis.  
Everything contributes a little. Nothing overwhelms.

Both approaches fight the same problem:  
a model — or a report — that is too heavily influenced  
by specific data points that don't generalise well.

## LinkedIn post idea

In D365, we spend days deciding which financial dimensions actually matter.  
Too many dimensions = noise. Reports that mean nothing.  
Too few = missing insight. Reports that hide problems.

In machine learning, this is solved automatically.  
It's called regularisation.  
The algorithm reviews every input feature and penalises the ones  
that are adding complexity without adding accuracy.

L1 throws away the irrelevant ones entirely.  
L2 keeps them all but reduces their influence.

The underlying problem is identical:  
too much detail obscures the signal.  
Whether you're configuring a chart of accounts or training an ML model —  
simplicity that preserves accuracy is the goal.
