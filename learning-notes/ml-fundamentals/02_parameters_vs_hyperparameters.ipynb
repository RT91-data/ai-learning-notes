## Parameters vs Hyperparameters

**Parameters** are values the model learns automatically during training.  
**Hyperparameters** are values YOU set before training begins.

## How it actually works

During training, the model adjusts its parameters thousands of times to reduce error.  
Hyperparameters control *how* that adjustment happens — you choose them upfront.  
Wrong hyperparameters = slow training or a model that never learns properly.

## Where it breaks or gets dangerous

**Wrong learning rate (hyperparameter):**  
Too high = model overshoots, weights bounce around, never converges.  
Too low = training takes forever or gets stuck before reaching best performance.

**Wrong number of layers or neurons (hyperparameter):**  
Too few = model is too simple to learn the pattern (underfitting).  
Too many = model memorises training data instead of learning it (overfitting).

**Wrong number of training epochs (hyperparameter):**  
Too few = model has not learned enough yet.  
Too many = model starts memorising noise in the training data.

**The core danger:**  
There is no formula for perfect hyperparameters.  
You tune them through experimentation — and a bad choice wastes days of compute time  
or silently produces a model that looks trained but performs poorly in production.

```python
# Parameters vs Hyperparameters — showing the relationship in action

import random

# HYPERPARAMETERS — you set these before training starts
learning_rate = 0.1
epochs = 10

# PARAMETER — the model starts with a random guess and learns the correct value
# Task: learn that the correct weight is approximately 100
weight = random.uniform(0, 1)  # random starting point
target = 100.0

print(f"Hyperparameters: learning_rate={learning_rate}, epochs={epochs}")
print(f"Parameter starting value (random guess): {weight:.4f}")
print(f"Target the model needs to learn: {target}")
print()

# Training loop — model adjusts the parameter each epoch
for epoch in range(epochs):
    error = weight - target                          # how wrong is the model?
    weight = weight - learning_rate * error          # adjust parameter to reduce error

    print(f"Epoch {epoch + 1}: weight = {weight:.4f}, error = {error:.4f}")

print(f"\nFinal learned parameter: {weight:.4f}")
print(f"You set the hyperparameters. The model learned the parameter.")
```

The actual transaction data the system processes and stores is like parameters —
