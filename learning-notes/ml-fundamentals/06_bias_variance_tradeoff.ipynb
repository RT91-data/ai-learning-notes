## Bias-Variance Tradeoff

**Bias:** model didn't learn enough from training data — underfitting.  
**Variance:** model learned training data too well, fails on new data — overfitting.  
The tradeoff: reducing one tends to increase the other.

## How it actually works

Total model error has three components:  
**Error = Bias² + Variance + Irreducible Noise**

**Irreducible noise** — random noise in data no model can eliminate. Ignore it.

**High Bias (Underfitting):**  
Model is too simple. Doesn't capture real patterns.  
Fails on training data AND new data.  
Like a consultant who applies the same solution to every client regardless of their situation.

**High Variance (Overfitting):**  
Model is too complex. Memorises training data including noise.  
Performs perfectly on training data, fails on new data.  
Like a consultant who memorises one client's setup and tries to apply it everywhere else.

**The sweet spot:**  
Complex enough to learn real patterns.  
Simple enough to generalise to new data.

## Where it breaks or gets dangerous

**The trap:** you optimise for training accuracy and declare success.  
Training accuracy: 97%. Test accuracy: 61%. You just shipped an overfitted model.

**High bias trap:**  
Model looks equally bad in testing and production.  
Easy to detect — both training and test accuracy are low.  
Hard to fix without fundamentally changing the model architecture.

**High variance trap:**  
Model looks great in testing if you tested on training data.  
Looks terrible in production on real unseen data.  
More dangerous because it's easy to miss during development.

**The rule:**  
Always evaluate on data the model never saw during training.  
Never celebrate training accuracy alone.

```python
# Bias-Variance Tradeoff — illustration

import random

def simulate_model(complexity, training_data, test_data):
    """
    Simulates how model complexity affects bias and variance.
    Higher complexity = lower bias, higher variance.
    Lower complexity = higher bias, lower variance.
    """
    # Simplified simulation
    if complexity == "too_simple":
        train_accuracy = 0.62
        test_accuracy = 0.60
        issue = "High Bias — Underfitting"
    elif complexity == "just_right":
        train_accuracy = 0.91
        test_accuracy = 0.88
        issue = "Balanced — Good generalisation"
    elif complexity == "too_complex":
        train_accuracy = 0.98
        test_accuracy = 0.61
        issue = "High Variance — Overfitting"
    
    return train_accuracy, test_accuracy, issue

training_data = list(range(100))
test_data = list(range(100, 130))

print(f"{'Complexity':<15} {'Train Acc':>10} {'Test Acc':>10}  Issue")
print("-" * 65)

for complexity in ["too_simple", "just_right", "too_complex"]:
    train_acc, test_acc, issue = simulate_model(complexity, training_data, test_data)
    print(f"{complexity:<15} {train_acc:>10.0%} {test_acc:>10.0%}  {issue}")

print()
print("Key insight: Watch the GAP between train and test accuracy.")
print("Large gap = overfitting. Both low = underfitting.")
```

## My D365 analogy

**High Bias — Underfitting:**  
A posting rule so generic it works for nothing properly.  
Example: one journal type for all transaction types across all legal entities.  
Too simple. Fails everywhere.

**High Variance — Overfitting:**  
A posting rule configured with 50 specific conditions for one legal entity.  
Works perfectly for that entity. Breaks for every other entity.  
Too specific. Doesn't generalise.

**The sweet spot:**  
A well-designed posting profile that covers the core patterns across entities  
with minor local overrides where genuinely needed.  
That's what a good D365 architect builds.  
That's also what a well-trained ML model does.
