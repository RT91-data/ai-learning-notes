## Dropout

A regularisation technique specific to neural networks.  
Randomly switches off a percentage of neurons during each training step.  
Forces every neuron to learn independently. Prevents overfitting.

## How it actually works

During training, at each step:  
- Randomly select a percentage of neurons (typically 20-50%)  
- Switch them off — they contribute nothing that round  
- Their weights don't update  
- Next step — different random neurons get switched off  

**Why does randomly breaking your network help it learn better?**

Without dropout, neurons start to co-depend.  
Neuron A learns that Neuron B will always correct its mistakes.  
So A gets lazy — stops learning robust patterns on its own.

With dropout, Neuron A can't rely on B — B might be off this round.  
Every neuron is forced to learn patterns useful on their own.  
Result: more robust network that generalises better to new data.

**Critical detail:**  
Dropout only happens during TRAINING.  
At inference (production), all neurons are active.  
You want the full network making real predictions.

**Dropout rate — a hyperparameter YOU set:**  
Typically 0.2 to 0.5 (20% to 50% of neurons dropped each round).  
Too high = too many neurons off, network can't learn. High bias.  
Too low = neurons still co-depend. Still overfitting. High variance.

## Where it breaks or gets dangerous

**Dropout rate too high:**  
Switching off 70-80% of neurons each round.  
Network loses too much information per step.  
Training becomes unstable. Model never converges properly.

**Dropout rate too low:**  
Not enough disruption to force independent learning.  
Neurons still develop co-dependencies.  
Overfitting persists despite using dropout.

**Applying dropout at inference:**  
A common implementation mistake.  
If dropout stays active during production predictions,  
results become random and inconsistent.  
Every call to the model gives a different answer for the same input.

**Dropout is not always the answer:**  
Works well for large neural networks with many parameters.  
Less effective for small networks or non-neural ML models.  
Don't apply it blindly to every architecture.

```python
# Dropout — conceptual illustration

import random

def apply_dropout(neurons, dropout_rate):
    """
    Simulates dropout during training.
    Randomly switches off neurons based on dropout rate.
    """
    result = []
    dropped = 0
    for i, neuron_value in enumerate(neurons):
        if random.random() < dropout_rate:
            result.append(0)  # neuron switched off
            dropped += 1
        else:
            result.append(neuron_value)
    return result, dropped

# Simulated neuron values in a layer
neuron_layer = [0.8, 0.3, 0.9, 0.1, 0.7, 0.5, 0.6, 0.2, 0.4, 0.85]
dropout_rate = 0.3  # hyperparameter — YOU set this

print(f"Original neuron layer ({len(neuron_layer)} neurons):")
print([round(n, 2) for n in neuron_layer])
print()

print("Training — dropout active:")
for step in range(4):
    dropped_layer, dropped_count = apply_dropout(neuron_layer, dropout_rate)
    print(f"  Step {step+1}: {[round(n, 2) for n in dropped_layer]}  ({dropped_count} neurons dropped)")

print()
print("Inference — dropout OFF, all neurons active:")
print(f"  {[round(n, 2) for n in neuron_layer]}  (full network, no dropout)")
print()
print(f"Dropout rate: {dropout_rate} — YOU set this hyperparameter before training")
```

## My D365 analogy

Imagine training a D365 implementation team  
where on any given day, random team members are unavailable.

Each consultant is forced to learn the full process end to end —  
posting logic, approval flows, dimension setup, period end closing —  
because they cannot rely on a specific colleague always being there.

Result: a more resilient team.  
Any member can handle any situation.  
No single point of failure.

Without dropout — one consultant owns posting logic, another owns approvals.  
Remove either one and the team collapses.  
Classic knowledge silo problem.

Dropout forces the neural network to avoid knowledge silos.  
Every neuron must be useful on its own.

Neural networks have the same problem.  
Neurons develop dependencies on each other.

Randomly remove neurons during training.  
Force every neuron to be useful on its own.
