## Transformers and Attention

Transformer is the architecture that powers every major LLM — Claude, GPT-4, Gemini.  
Attention is the key innovation inside transformers.  
It lets the model focus on the most relevant parts of the input when generating each word.

## How it actually works

**The core problem LLMs solve:**  
Given a sequence of words, predict the next most likely word.  
Everything Claude does — answering questions, writing code, explaining concepts —  
is next-word prediction done billions of times at extreme sophistication.

**What is a Transformer:**  
A neural network architecture designed specifically for sequential data like text.  
Before transformers, models processed text word by word in order (slow, forgot early context).  
Transformers process all words simultaneously and use attention to connect them.

**What is Attention:**  
When predicting the next word, not all previous words are equally important.  

Example:  
"The invoice was submitted by the vendor but it was rejected because ___"  
To predict what comes next, the model needs to focus on "rejected" and "invoice."  
Not on "the", "but", "by."  
Attention assigns different weights to different words based on relevance.

**Three components of attention:**  
- **Query:** what am I looking for right now?  
- **Key:** what does each word offer?  
- **Value:** what information does each word actually contain?  
Attention computes similarity between query and all keys,  
uses that to weight the values — focusing on what matters most.

**Why transformers scale better than previous architectures:**  
Attention looks at all words simultaneously.  
Previous models (RNNs) processed word by word — slow, forgot early context.  
Transformers parallelise — faster training, better at long documents.

## Where it breaks or gets dangerous

**Attention complexity:**  
Attention computes relationships between every word and every other word.  
Cost grows quadratically with sequence length — O(n²).  
Double the input length = four times the compute cost.  
This is why very long contexts are expensive.

**Lost in the middle:**  
Attention weakens for content in the middle of very long contexts.  
Models pay more attention to the beginning and end of the input.  
Practical implication: put your most important instructions  
at the start or end of your prompt, not buried in the middle.

**Hallucination still happens:**  
Transformers predict likely next tokens — not factually correct ones.  
Attention helps the model be coherent and contextually aware.  
It does not guarantee factual accuracy.

**Transformer ≠ model:**  
Transformer is the architecture — the blueprint.  
Claude, GPT-4, Gemini are models built on transformer architecture.  
Same family, different buildings.

```python
# Transformers and Attention — conceptual illustration
# Simplified attention weight calculation

import math

def softmax(scores):
    """Converts raw scores to probabilities that sum to 1"""
    exp_scores = [math.exp(s) for s in scores]
    total = sum(exp_scores)
    return [e / total for e in exp_scores]

def simplified_attention(query_word, words, relevance_scores):
    """
    Simplified attention: shows how much each word is attended to
    when predicting the next token after query_word.
    In reality, scores are computed via dot products of learned vectors.
    """
    attention_weights = softmax(relevance_scores)
    return attention_weights

# Example: predicting next word after "rejected"
# Sentence: "The invoice was submitted by vendor but it was rejected because"
words = ["The", "invoice", "was", "submitted", "by", "vendor", "but", "it", "was", "rejected", "because"]

# Relevance scores — how relevant each word is when processing "rejected"
# In real transformers these are learned during training
relevance_scores = [0.1, 2.5, 0.2, 0.8, 0.1, 1.8, 0.1, 0.3, 0.2, 3.0, 1.5]

attention_weights = simplified_attention("rejected", words, relevance_scores)

print("Attention weights when processing 'rejected':")
print()
print(f"{'Word':<12} {'Attention Weight':>18} {'Bar'}")
print("-" * 50)
for word, weight in zip(words, attention_weights):
    bar = "█" * int(weight * 50)
    print(f"{word:<12} {weight:>18.4f}  {bar}")

print()
print("High attention = model focuses here when predicting next word.")
print("'rejected', 'invoice', 'vendor' get highest attention — makes sense.")
print("'The', 'but', 'by' get almost none.")
```

## My D365 analogy

During a D365 financial audit, you don't read every transaction  
with equal attention.

You focus on:  
- High value entries  
- Unusual vendors  
- Late postings  
- Transactions with missing dimensions  

You reduce attention on:  
- Small routine transactions  
- Known recurring entries  
- Transactions from long-established vendors  

That selective focus is exactly what attention does.  
It concentrates on what matters most for the current question.  
It reduces weight on what is routine and irrelevant.

A senior D365 auditor and a transformer model  
use the same cognitive strategy —  
focus where the signal is strongest.
