## Types of Machine Learning

Three types: Supervised, Unsupervised, Reinforcement.  
Most enterprise AI you will ever build is supervised.

## Supervised Learning

You provide data with correct answers (labels).  
Model learns to map input → correct output.

Think of it as learning with an answer key.  
Every training example has a question AND the right answer.  
The model adjusts until its answers match yours.

**When to use it:** whenever you have historical data with known outcomes.  
Examples: spam/not spam, fraud/not fraud, approved/rejected, routed to manager A or B.

## Unsupervised Learning

You provide data with NO labels. No correct answers.  
Model finds hidden structure and patterns on its own.

Think of it as handing someone 10,000 documents with no index.  
They group similar ones together without being told what "similar" means.

**When to use it:** when you do not know what you are looking for yet,  
or when labelling data is impossible or too expensive.  
Examples: customer segmentation, vendor anomaly detection, topic clustering.

## Reinforcement Learning

Model learns by trial and error.  
It takes actions, receives rewards for correct ones, penalties for wrong ones.  
Over time it learns the strategy that maximises reward.

**When to use it:** sequential decision making, game playing, robotics.  
Rarely used directly in enterprise ERP AI — but the concept matters for AI agents.  
When you build an AI agent that autonomously processes D365 workflows,  
reinforcement learning principles are underneath it.

## The critical distinction most people get wrong

Unsupervised ≠ "the model decides on its own"  
Unsupervised = no labels exist in the training data

If you have historical data with known outcomes — it is supervised, always.  
Even if the model makes autonomous predictions later.

Example trap:  
"Approval routing should be unsupervised because the model decides the route."  
Wrong. You have years of routing history with known outcomes. That is a labelled dataset.  
Train a supervised classifier. It will dramatically outperform an unsupervised approach.

```python
# Supervised vs Unsupervised — conceptual illustration

# SUPERVISED: data has labels
invoices_supervised = [
    {"amount": 50000, "vendor_age_days": 30, "label": "fraudulent"},
    {"amount": 1200, "vendor_age_days": 800, "label": "legitimate"},
    {"amount": 98000, "vendor_age_days": 15, "label": "fraudulent"},
    {"amount": 450, "vendor_age_days": 1200, "label": "legitimate"},
]
print("Supervised — model learns from labels:")
for inv in invoices_supervised:
    print(f"  Amount: {inv['amount']}, Vendor age: {inv['vendor_age_days']} days → {inv['label']}")

print()

# UNSUPERVISED: same data, no labels
invoices_unsupervised = [
    {"amount": 50000, "vendor_age_days": 30},
    {"amount": 1200, "vendor_age_days": 800},
    {"amount": 98000, "vendor_age_days": 15},
    {"amount": 450, "vendor_age_days": 1200},
]
print("Unsupervised — model finds clusters on its own, no labels:")
for inv in invoices_unsupervised:
    print(f"  Amount: {inv['amount']}, Vendor age: {inv['vendor_age_days']} days → ???")
print("  Model might group high-amount + new-vendor together — but YOU decide if that means fraud")
```

```python
## My D365 analogy

Supervised = D365 workflow with a complete audit trail.  
Every transaction has a known outcome — posted, rejected, escalated.  
That history is your labelled dataset. Use supervised learning.

Unsupervised = a brand new D365 implementation with zero transaction history.  
No labels exist yet. You use unsupervised to find natural groupings in the data  
before you know what questions to ask.

Most experienced D365 consultants are sitting on years of supervised learning gold  
and do not realise it.
```
