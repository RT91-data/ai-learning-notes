## Overfitting and Underfitting

**Overfitting:** model performs well on training data, fails on new unseen data.  
**Underfitting:** model performs poorly on both training data and new data.

## How it actually works

Before training, data is split into two sets:  
- **Training set:** what the model learns from  
- **Test set:** data the model has never seen — used to check real performance

**Overfitting:**  
The model memorises the training set instead of learning the underlying pattern.  
Give it the same questions it trained on — perfect answers.  
Give it a new question — it fails.  
Like a student who memorised every past exam paper word for word.  
Change one question — they collapse. They never learned the subject.

**Underfitting:**  
The model did not learn enough from the training data.  
It fails on training questions AND new questions.  
Like a student who did not study at all.

## Where it breaks or gets dangerous

**Overfitting is the more dangerous failure in enterprise AI:**  
The model looks great during testing — because you tested it on training data.  
You deploy it to production. It fails on real transactions. Nobody saw it coming.

**How to detect overfitting:**  
Training accuracy: 96%  
Test accuracy: 61%  
That gap is the red flag. The bigger the gap, the worse the overfitting.

**How to detect underfitting:**  
Training accuracy: 58%  
Test accuracy: 55%  
Both are low. The model never learned anything useful.

**The trap:**  
Never evaluate a model only on training data.  
Always hold back a test set the model never sees during training.  
If you skip this step, you will deploy a broken model with confidence.

```python
# Overfitting vs Underfitting — conceptual illustration
# Simulating model performance on training vs test data

results = {
    "Healthy model": {
        "training_accuracy": 0.91,
        "test_accuracy": 0.88,
        "verdict": "Good — small gap, generalises well"
    },
    "Overfitted model": {
        "training_accuracy": 0.97,
        "test_accuracy": 0.61,
        "verdict": "Danger — memorised training data, fails on new data"
    },
    "Underfitted model": {
        "training_accuracy": 0.58,
        "test_accuracy": 0.55,
        "verdict": "Danger — never learned the pattern at all"
    }
}

print(f"{'Model':<25} {'Train Acc':>10} {'Test Acc':>10}  Verdict")
print("-" * 75)
for model, scores in results.items():
    print(f"{model:<25} {scores['training_accuracy']:>10.0%} "
          f"{scores['test_accuracy']:>10.0%}  {scores['verdict']}")
```

## My D365 analogy

Imagine training an invoice posting model on historical data  
where every invoice used the same GL accounts, cost centres, and dimensions.

**Overfitting:**  
A new invoice arrives with a dimension combination the model has never seen.  
It cannot post it correctly — or posts it to the wrong account entirely.  
The model learned the specific combinations, not the posting logic.

**Underfitting:**  
The model was trained on too little data or was too simple.  
It cannot even handle invoices it has already seen in training.  
Completely unusable in production.

In D365 terms: an overfitted model is like a posting setup  
configured for one legal entity and blindly copied to all others.  
Works perfectly for the original. Breaks everywhere else.

## LinkedIn post idea

In D365, we test configurations in UAT before go-live.  
But here is what most people do wrong —  
they test with the same data they used to build the configuration.

Of course it works. It was built for exactly that data.

Machine learning has the same trap. It is called overfitting.  
The model looks perfect during testing because you tested it on training data.  
You go live. Real transactions come in. It fails.

The fix is simple: always test on data the model has never seen.  
In D365 terms — UAT with data from a different entity, a different period, a different scenario.  
If it breaks there, it was never ready for production.
