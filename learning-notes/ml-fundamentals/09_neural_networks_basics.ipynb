## Neural Networks Basics

A neural network is a series of mathematical functions stacked in layers.  
It learns by adjusting millions of internal numbers (weights) until its predictions are accurate.  
"Deep" learning simply means more than one hidden layer.

## How it actually works

**Three layers:**

**Input layer:**  
Receives raw data.  
Each input feature becomes one neuron.  
Example: analysing an invoice → inputs could be amount, vendor age, line item count, country code.

**Hidden layers:**  
Where learning happens.  
Each neuron takes inputs from the previous layer,  
multiplies them by weights, adds a bias value, passes result forward.  
Multiple hidden layers = "deep" learning.  
Each layer extracts more complex patterns from the previous layer's output.

**Output layer:**  
Produces the final answer.  
Fraud or not fraud. Spam or not spam. Which approval route.

**How it learns — three steps repeated thousands of times:**  
1. Forward pass: input flows through all layers, model makes a prediction  
2. Cost function measures how wrong the prediction is  
3. Backpropagation: error signal travels backwards through layers,  
   weights adjusted slightly to reduce error  
   
**Backpropagation** — the word you will hear in every ML interview.  
It is simply gradient descent applied backwards through the network layers.

## Where it breaks or gets dangerous

**Vanishing gradients:**  
In deep networks, the error signal gets smaller as it travels backwards through layers.  
Early layers receive almost no signal — they stop learning entirely.  
The network has many layers but only the last few are actually improving.  
Solutions: better activation functions (ReLU), residual connections, batch normalisation.

**Too many layers, too little data:**  
Deep networks need large amounts of data to train well.  
With small datasets, a deep network overfits severely.  
A simpler model often outperforms a deep network on small enterprise datasets.

**Black box problem:**  
Neural networks are extremely difficult to interpret.  
You cannot easily explain WHY the model made a specific decision.  
Critical problem in enterprise and regulated industries —  
a CFO or auditor will ask why an invoice was flagged as fraudulent.  
A neural network cannot easily tell you.

**Compute cost:**  
Large networks are expensive to train.  
Fine-tuning on enterprise data requires careful cost management.

```python
# Neural Network — simplified conceptual illustration
# Single hidden layer, manual forward pass

import math

def relu(x):
    """Activation function — introduces non-linearity"""
    return max(0, x)

def sigmoid(x):
    """Output activation — squashes value between 0 and 1"""
    return 1 / (1 + math.exp(-x))

def forward_pass(inputs, hidden_weights, output_weights):
    """
    Simplified forward pass through a neural network.
    Input layer -> Hidden layer -> Output layer
    """
    # Hidden layer
    hidden_outputs = []
    for weights in hidden_weights:
        raw = sum(i * w for i, w in zip(inputs, weights))
        activated = relu(raw)
        hidden_outputs.append(activated)
    
    # Output layer
    raw_output = sum(h * w for h, w in zip(hidden_outputs, output_weights))
    final_output = sigmoid(raw_output)
    
    return hidden_outputs, final_output

# Invoice fraud detection example
# Inputs: [amount_normalised, vendor_age_days_normalised, line_items_normalised]
invoice_inputs = [0.85, 0.12, 0.90]  # high amount, new vendor, many line items

# Pretrained weights (in reality these are learned via backpropagation)
hidden_weights = [
    [0.5, -0.3, 0.8],   # neuron 1
    [-0.2, 0.9, 0.1],   # neuron 2
    [0.7, 0.4, -0.5],   # neuron 3
]
output_weights = [0.6, -0.4, 0.8]

hidden_out, prediction = forward_pass(invoice_inputs, hidden_weights, output_weights)

print("Invoice Neural Network — Forward Pass")
print(f"Inputs (amount, vendor age, line items): {invoice_inputs}")
print(f"Hidden layer outputs: {[round(h, 3) for h in hidden_out]}")
print(f"Final prediction (fraud probability): {prediction:.3f}")
print()
if prediction > 0.5:
    print("Result: FLAGGED AS FRAUDULENT")
else:
    print("Result: Legitimate invoice")
print()
print("In real training, weights are adjusted via backpropagation")
print("until predictions match actual fraud labels in training data.")
```

## My D365 analogy

A neural network processes information in layers —  
just like a D365 multi-level approval workflow.

**Input layer** = invoice arrives with all its attributes:  
amount, vendor, legal entity, cost centre, purchase category.

**Hidden layer 1** = first approval level checks basic rules:  
is the amount within budget? Is the vendor approved?

**Hidden layer 2** = second level checks risk patterns:  
is this vendor new? Is the amount unusually high for this category?

**Output layer** = final decision:  
approve, reject, or escalate to senior management.

Each layer adds information to the decision.  
Each layer builds on what the previous layer found.  
The final decision emerges from the combined processing of all layers.

The difference: in D365 you configure every rule manually.  
In a neural network, the rules are learned automatically from historical data.
