## Hallucination and Grounding

Hallucination = the model generates confident, plausible-sounding text that is factually wrong.  
Grounding = giving the model factual context to anchor its responses to real information.  
RAG is the primary grounding technique for enterprise AI systems.

## How it actually works

**Why hallucination happens:**  
LLMs predict the next most likely token based on patterns learned during training.  
The model has no concept of "I don't know."  
When it hits a knowledge gap, it fills it with statistically likely text —  
not factually correct text.  
It sounds confident because confident-sounding text is what it learned to produce.

**Three types of hallucination:**  
- **Factual hallucination:** states wrong facts, dates, numbers, names  
- **Source hallucination:** invents citations, paper titles, URLs that don't exist  
- **Reasoning hallucination:** logical steps sound correct but conclusion is wrong  

**What grounding means:**  
Providing the model with real, verified source material  
so it generates answers based on actual documents  
rather than learned patterns alone.

**How RAG grounds the model:**  
1. User asks a question  
2. System retrieves relevant document chunks  
3. Chunks are injected into the prompt  
4. Model generates answer based on injected content  
5. Model cites the source chunks  

The model now has real information to work from.  
It doesn't need to rely on memory. It reads and summarises.

## Where it breaks or gets dangerous

**Hallucination in enterprise AI is not just embarrassing — it's dangerous:**  
Wrong GL account recommendation → incorrect financial posting  
Wrong tax code → compliance violation  
Wrong approval threshold → unauthorised transaction  
Invented regulatory reference → legal risk  

**Grounding doesn't eliminate hallucination:**  
Even with retrieved context, the model can still hallucinate.  
It may ignore the context and generate from memory.  
It may misread or misinterpret the retrieved content.  
It may blend retrieved content with incorrect remembered facts.  

**The confidence problem:**  
Hallucinated answers sound exactly as confident as correct answers.  
Users trust confident-sounding AI output by default.  
In enterprise settings this is a serious risk.  

**When grounding fails:**  
- Retrieved chunks are wrong or irrelevant (retrieval problem)  
- Source documents are outdated (data freshness problem)  
- System prompt doesn't instruct the model to stick to context  
- Temperature is too high — model gets "creative"

```python
# Hallucination and Grounding — conceptual illustration

def generate_answer(question, context=None, temperature="high"):
    """
    Simulates how grounding affects answer quality.
    Without context: model relies on memory, risk of hallucination.
    With context: model anchors to provided information.
    """
    
    if context is None:
        # No grounding — model relies on training memory
        if temperature == "high":
            answers = {
                "What is the vendor payment term for Contoso?": 
                    "Based on standard practice, Contoso typically operates on NET 30 payment terms.",  # hallucinated
                "What was the Q3 invoice approval rate?":
                    "The Q3 invoice approval rate was approximately 94.3%.",  # hallucinated number
            }
        return answers.get(question, "Answer generated from memory — may be inaccurate."), "NO GROUNDING"
    
    else:
        # Grounded — model uses provided context
        return f"Based on the provided document: {context}", "GROUNDED"

# Test without grounding
question1 = "What is the vendor payment term for Contoso?"
answer1, source1 = generate_answer(question1)
print(f"Question: {question1}")
print(f"Source: {source1}")
print(f"Answer: {answer1}")
print()

# Test with grounding
context = "Contoso vendor agreement dated 2024-01-15: Payment terms NET 60. Early payment discount 2% if paid within 10 days."
answer2, source2 = generate_answer(question1, context=context, temperature="low")
print(f"Question: {question1}")
print(f"Source: {source2}")
print(f"Answer: {answer2}")
print()

print("="*60)
print("Key insight:")
print("Without grounding: model invents plausible-sounding answer.")
print("With grounding: model anchors to real document content.")
print("Same question. Very different reliability.")
print()
print("In enterprise D365 AI — always ground with real data.")
```

## My D365 analogy

In D365, a junior consultant who doesn't know the answer  
has two choices:  
1. Say "I don't know, let me check the configuration"  
2. Make up a confident-sounding answer based on what usually works  

Option 2 is hallucination.  
It happens in consulting. It causes expensive mistakes.  

A grounded AI assistant is like a consultant  
who always checks the actual system configuration before answering.  
"According to the posting profile setup in legal entity USMF..."  
Not "based on standard practice..."  

The difference between a dangerous AI assistant  
and a trustworthy one is grounding.  
Always anchor responses to real, verified source data.
