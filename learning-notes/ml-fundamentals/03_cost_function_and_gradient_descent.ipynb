## Cost Function and Gradient Descent

**Cost function** measures how wrong the model's predictions are right now.  
**Gradient descent** is the process the model uses to become less wrong, step by step.

## How it actually works

**Cost function:**  
The model makes a prediction. The cost function compares that prediction to the correct answer.  
Big gap = high cost. Perfect prediction = zero cost.  
Think of it as a score that measures failure, not success. Lower is better.

**Gradient descent:**  
Imagine you are blindfolded on a hilly landscape. Your goal is the lowest point.  
You cannot see the whole map. So you feel the ground under your feet,  
find which direction slopes downward, and take one small step that way. Then repeat.  
Each step = the model adjusting its weights slightly to reduce cost.  
After thousands of steps, the model reaches the lowest cost it can find.

**Learning rate** controls how big each step is.  
Too large = you overshoot the lowest point and bounce around forever.  
Too small = you get there eventually but training takes forever.

## Where it breaks or gets dangerous

**Local minima:**  
The blindfolded hiker might reach a small valley and think they are at the bottom —  
but the actual lowest point is somewhere else on the map.  
The model gets stuck in a "good enough" solution, not the best one.

**Learning rate too high:**  
Model overshoots the lowest point repeatedly. Cost never settles. Training fails.

**Learning rate too low:**  
Training takes so long it becomes impractical.  
Or the model gets stuck and stops improving before reaching the bottom.

**Vanishing gradients:**  
In deep neural networks, the gradient signal gets so small as it travels back through layers  
that early layers stop learning entirely. A known failure mode in deep learning.

```python
# Cost Function and Gradient Descent — simple illustration
# Linear regression: predict invoice amount from number of line items

# Training data: [line_items, actual_amount]
data = [(2, 200), (5, 480), (3, 310), (8, 790), (1, 105)]

# Start with a random weight (parameter)
weight = 0.0
learning_rate = 0.01  # hyperparameter — YOU set this

# Simple gradient descent loop
for epoch in range(100):
    total_cost = 0
    gradient = 0

    for line_items, actual_amount in data:
        prediction = weight * line_items
        error = prediction - actual_amount
        total_cost += error ** 2          # cost = squared error
        gradient += 2 * error * line_items  # direction to reduce cost

    weight = weight - learning_rate * (gradient / len(data))  # take one step

    if epoch % 20 == 0:
        print(f"Epoch {epoch}: weight = {weight:.2f}, cost = {total_cost:.2f}")

print(f"\nFinal weight: {weight:.2f}")
print(f"Prediction for 6 line items: ${weight * 6:.2f}")
```
