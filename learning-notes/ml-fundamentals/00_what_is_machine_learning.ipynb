## What is Machine Learning?

Teaching a computer to learn from examples instead of programming it with explicit rules.

## How it actually works

Traditional programming: YOU write the rules. Computer follows them.  
Machine Learning: YOU give examples with correct answers. Computer figures out the rules itself.

Three steps in every ML problem:
1. Feed the model data (examples)
2. Model finds patterns in that data
3. Model uses those patterns to make predictions on new, unseen data

The model is never told the rules. It discovers them.

## A simple real-world example

You want a computer to identify spam emails.

Traditional approach: write rules manually.
- If email contains "free money" → spam
- If email contains "click here" → spam
- Problem: spammers change words. Your rules break instantly.

ML approach: show the computer 100,000 emails labelled spam or not spam.
- Computer finds its own patterns across all 100,000 examples
- It catches patterns you would never think to write manually
- When spammers change words, you retrain with new examples — done.

## Three types of Machine Learning

| Type | What it means | Example |
|------|--------------|---------|
| Supervised | Learn from labelled examples | Spam filter, fraud detection |
| Unsupervised | Find hidden patterns, no labels | Customer segmentation, anomaly detection |
| Reinforcement | Learn by trial and error, get rewarded for correct actions | Game playing AI, robotics |

Most enterprise AI you will build is supervised learning.

```python
# Machine Learning in 10 lines — conceptual illustration
# No real training here, just showing the idea

# Step 1: Your data (examples with correct answers)
emails = [
    {"text": "free money click here", "label": "spam"},
    {"text": "meeting at 3pm tomorrow", "label": "not spam"},
    {"text": "you won a prize claim now", "label": "spam"},
    {"text": "project update attached", "label": "not spam"},
]

# Step 2: In real ML, a model would learn patterns from this data
# Step 3: Then predict on a new email it has never seen
new_email = "congratulations you have been selected"

# A trained model would output:
print(f"Email: '{new_email}'")
print(f"Prediction: spam")  # because it learned the pattern
print()
print("The model learned this — you never told it 'congratulations = spam'")
```
